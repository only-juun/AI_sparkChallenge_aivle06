```python
a_label = a['label']
```

```python
cnt = 0
for i in range(len(a_label)):
    if a_label[i] != x_label[i]:
        cnt += 1
        print(i)
        print(f'a = {a_label[i]}', end = ' ')
        print(f'x = {x_label[i]}', end = ' ')
        print()
print()
print(cnt)
```
